```python
y = df['diagnosis']
x = df.drop(['diagnosis'], axis=1)
```

```python
x.head()
```

```python
B, M = y.value_counts()
print('Number of Benign:', B)
print('Number of Malignant:', M)

# plotting a countplot
fig = sns.countplot(y, label='count')
```

```python
x.describe()
```

```python
correlation = x.corr()

# plotting a heatmap
plt.figure(figsize=(20,20))
hm = sns.heatmap(correlation, annot=True)
plt.show()
```
