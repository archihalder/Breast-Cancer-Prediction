```python
df = df.drop(['perimeter_mean','area_mean','compactness_mean', 'concave points_mean',
              'perimeter_se', 'area_se'], axis=1)
```

```python
df.shape
```

```python
# splitting the dataset
X = df.iloc[:,1:].values
Y = df.iloc[:,0].values
```

```python
type(X)
```

```python
X.shape, Y.shape
```

```python
# splitting the independent dataset
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_test = ss.fit_transform(X_test)
```

```python
X_train
```

### Creating and Training models

```python
# logistic regression 
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train, Y_train)
```

```python
# accuracy of logistic regression model
log_reg_acc = log_reg.score(X_train, Y_train)
print(f'Logistic Regression Accuracy: {log_reg_acc*100}%')
```

```python
# decision tree
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(criterion='entropy')
dt.fit(X_train, Y_train)
```

```python
# accuracy of logistic regression model
dt_acc = dt.score(X_train, Y_train)
print(f'Decision Tree Accuracy: {dt_acc*100}%')
```

```python
# random forest classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=10, criterion='entropy')
rf.fit(X_train, Y_train)
```

```python
# accuracy of logistic regression model
rf_acc = rf.score(X_train, Y_train)
print(f'Random Forest Classifier Accuracy: {rf_acc*100}%')
```

### Testing the Accuracy of the Models

```python
from sklearn.metrics import confusion_matrix

def tester(model):
    cm = confusion_matrix(Y_test, model.predict(X_test))
    
    TP = cm[0][0]
    TN = cm[1][1]
    FN = cm[1][0]
    FP = cm[0][1]
    
    test_acc = ((TP + TN)/(TP + TN + FN + FP))*100
    fig = sns.heatmap(cm, annot=True)
    print('Testing Accuracy: ', test_acc)
    return fig    
```

```python
# logistic regression testing accuracy
tester(log_reg)
```

```python
# decision tree testing accuracy
tester(dt) 
```

```python
# random forest classifier testing accuracy
tester(rf)
```

Logistic Regression Model was found to be the most accurate while testing the model
